```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import random

# Configure options and attach to an existing Chrome session (ensure your debugger port matches)
options = Options()
options.debugger_address = "127.0.0.1:9222"  # Make sure this port is the same as your running Chrome instance
driver = webdriver.Chrome(options=options)

# Navigate to the target webpage (replace with the actual URL)
driver.get("https://app.apollo.io/#/people")

# Set up an explicit wait for dynamic content
wait = WebDriverWait(driver, 10)


```

```python

def human_pause(min_seconds=1, max_seconds=4):
    """Pause for a random duration to simulate human interaction."""
    delay = random.uniform(min_seconds, max_seconds)
    time.sleep(delay)

def click_on_job_titles_button(wait):
    """
    Click on the "Job Titles" button.
    The function waits until the element is clickable, simulates a human pause, and then clicks.
    """
    print("Looking for the Job Titles button...")
    # Locate the button element that contains a <span> with the text 'Job Titles'
    
    element = wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//span[text()='Job Titles']")
    ))
    
    human_pause()
    element.click()
    print("Job Titles button clicked.")
    human_pause()

def click_on_select_job_title(wait, job_title):
    
    """
    Click on the job title selection field and type in the provided job title.
    Mimics human interaction by clicking twice and pausing between actions.
    """
    print(f"Selecting job title: {job_title}")
    try:
        element = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//div[@class='Select-placeholder' and contains(text(), 'Search for a job title')]")
        ))
        element.click()
        
        human_pause(0.5, 1.5)  # a shorter pause for the double click
    except Exception as e:
        print("moving on to the next step")


        input_element = wait.until(EC.visibility_of_element_located(
            (By.XPATH, "//div[contains(@class, 'zp-select-main')]//input[contains(@class, 'Select-input')]")
        ))
        input_element.click()
        human_pause()
        clear_element = wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//span[contains(@class, 'Select-clear-zone')]")
        ))
        clear_element.click()
    input_element = wait.until(EC.visibility_of_element_located(
            (By.XPATH, "//div[contains(@class, 'zp-select-main')]//input[contains(@class, 'Select-input')]")
        ))
    human_pause()
    input_element.send_keys(job_title)
    human_pause(3, 5) # a longer pause to mimic thinking/typing
    input_element.send_keys(Keys.ENTER)
    
    input_element.send_keys(Keys.ENTER)
    
    print("Job title entered and submitted.")
    human_pause()

def click_on_companies_button(wait):
    """
    Click on the Companies (or Company) button on the page.
    """
    print("Looking for the Companies button...")
    element = wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//span[text()='Company']")
    ))
    human_pause()
    element.click()
    print("Companies button clicked.")
    human_pause()

def click_on_select_company(wait, company):
    """
    Click on the company selection field, type in the company name, and select it.
    Includes human-like pauses to simulate natural typing and selection.
    """
    print(f"Selecting company: {company}")
    try:
        element = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//div[@class='Select-placeholder' and contains(text(), 'Enter companies...')]")
        ))
        element.click()
        human_pause(0.5, 1.5)
    except Exception as e:
        print("moving on to the next step")

    

        input_element = wait.until(EC.visibility_of_element_located(
            (By.XPATH, "//div[contains(@class, 'zp-select-main')]//input[contains(@class, 'Select-input')]")
        ))
        
        input_element.click()
        human_pause()
        clear_element = wait.until(EC.element_to_be_clickable(
        (By.XPATH, "//span[contains(@class, 'Select-clear-zone')]")
        ))
        clear_element.click()
    input_element = wait.until(EC.visibility_of_element_located(
            (By.XPATH, "//div[contains(@class, 'zp-select-main')]//input[contains(@class, 'Select-input')]")
        ))
    human_pause()
    input_element.send_keys(company)
    # A longer pause to mimic thinking/typing before selection
    time.sleep(random.uniform(3, 5))
    human_pause(3, 5)

    # Press the DOWN arrow key to highlight the first option in the dropdown
    input_element.send_keys(Keys.DOWN)
    # Then press ENTER to select it
    input_element.send_keys(Keys.ENTER)
    human_pause()

def click_access_email_and_copy(wait, indx):
    """
    For a given table row (based on index), click the 'Access email' button, then copy and return details.
    Mimics human delays in clicking and reading the page.
    """
    # Locate the table row by its id
    print(f"Processing row with index {indx}...")
    row = wait.until(EC.presence_of_element_located((By.ID, f"table-row-{indx}")))
    
    try:
        access_email_button = wait.until(EC.element_to_be_clickable(
            (By.XPATH, "//button[.//span[contains(text(),'Access email')]]")
        ))
        human_pause()
        access_email_button.click()
        print("Access email button clicked.")
    except Exception as e:
        print("Error clicking 'Access email' button:", e)
    
    # Let the email content load naturally
    human_pause()
    
    email_element = wait.until(lambda driver: row.find_element(By.XPATH, "./div[4]//a"))
    email_address = email_element.text
    
    name_element = row.find_element(By.XPATH, "./div[1]//a")
    name = name_element.text.strip()
    
    role_element = row.find_element(By.XPATH, "./div[2]//span")
    role = role_element.text.strip()
    
    company_element = row.find_element(By.XPATH, "./div[3]//span")
    company = company_element.text.strip()
    
    print(f"Extracted data: {name} | {role} | {company} | {email_address}")
    human_pause()
    
    return {"name": name, "role": role, "company": company, "email": email_address}
```

```python

# Example usage:
company, job_title = "Oracle", "Manager"

def find_emails(company, job_title, wait):
    # Click on the Job Titles button and select the job title
    click_on_job_titles_button(wait)
    click_on_select_job_title(wait, job_title)
    driver.implicitly_wait(10)

    # Click on the Companies button and select the company
    click_on_companies_button(wait)
    click_on_select_company(wait, company)

# find_emails(company, job_title, wait)
```

```python
import random
# Loop to iterate through pages and collect email data
combined_data = []

job_titles = ["Senior Recruiter", "Recruiter", "Project Manager", "Senior software engineer"]
company= "Google"  # Example job titles
for job_title in job_titles:
    count_emails = 0
    page_count = 0
    find_emails(company, job_title, wait)
    while True:
        for i in range(0, 25):
            try : 
                wait.until(EC.presence_of_element_located((By.ID, f"table-row-{i}")))
            except Exception as e:
                break
            # Process each row and store the data
            data = click_access_email_and_copy(wait, i)
            combined_data.append(data)
            if data["email"] != "No email":
                count_emails += 1
            if count_emails > 15 or page_count >5: break

            if i%3 == 0: human_pause(1, 2)
        page_count += 1
        
        # Wait before moving to the next page to mimic a human reading the data
        human_pause(2, 4)
        
        if count_emails > 15 or page_count >5: break
        print("Moving to the next page...")
        try : 
            button = driver.find_element(By.CSS_SELECTOR, 'button[aria-label="Next"]')
            button.click()
        except Exception as e:
            print("No next button found or not clickable.")
            break


# # Write the combined data to a CSV file
# import csv

# csv_file_name = "combined_data.csv"
# with open(csv_file_name, mode='w', newline='', encoding='utf-8') as file:
#     writer = csv.DictWriter(file, fieldnames=["name", "role", "company", "email"])
#     writer.writeheader()
#     writer.writerows(combined_data)

# print(f"Data has been written to {csv_file_name}")
```

```python
# Write the combined data to a CSV file
import csv
import os

csv_file_name = f"{company}_data.csv"

# Check if the file already exists
file_exists = os.path.isfile(csv_file_name)

# Read existing emails from the file if it exists
existing_emails = set()
if file_exists:
    with open(csv_file_name, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            existing_emails.add(row['email'])

# Filter out rows with duplicate emails
filtered_data = [row for row in combined_data if row['email'] not in existing_emails]

# Open the file in append mode if it exists, otherwise create a new file
with open(csv_file_name, mode='a' if file_exists else 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["name", "role", "company", "email"])
    if not file_exists:
        writer.writeheader()  # Write header only if the file is new
    writer.writerows(filtered_data)

print(f"Data has been written to {csv_file_name}")
```
